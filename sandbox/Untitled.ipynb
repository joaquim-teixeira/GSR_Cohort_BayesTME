```python
from bayestme import spatial_expression


n_nodes = cleaned_dataset.n_spot_in
n_signals = cleaned_dataset.n_gene
n_spatial_patterns=10
alpha = np.ones(n_spatial_patterns + 1)
alpha[0] = 10
alpha[1:] = 1 / n_spatial_patterns
prior_vars = np.repeat(100.0, 2)


sde = spatial_expression.SpatialDifferentialExpression(
        n_cell_types=deconvolution_result.n_components,
        n_spatial_patterns=n_spatial_patterns,
        n_nodes=n_nodes,
        n_signals=n_signals,
        edges=cleaned_dataset.edges,
        alpha=alpha,
        prior_vars=np.repeat(100.0, 2),
        lam2=100
)
```
