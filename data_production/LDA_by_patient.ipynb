```python
import gensim
import sys
import urllib.request
import numpy as np
sys.path.insert(0, '/Users/joaquimteixeira/GSR_22_Tansey_Banerjee/bayestme-main/src')
from bayestme import data
from sklearn.model_selection import train_test_split
from sklearn import decomposition
import pandas as pd
from sklearn import decomposition
from bayestme import gene_filtering
import os 
import h5py 

fls=os.listdir('/Users/joaquimteixeira/GSR_22_Tansey_Banerjee/teixeira_code/flipped_cohort_data/')
fls=sorted(fls)

fout=[]
for k in fls:
    fout.append (k[:2])
n_comp = [6,6,7,6,5,4,7,5]

fls=os.listdir('/Users/joaquimteixeira/GSR_22_Tansey_Banerjee/teixeira_code/flipped_cohort_data/')
patients=['A','B','C','D','E','F','G','H']

```

```python
p=0
for pat in patients:
    print (pat)
    fls_pat= [k for k in fls if pat in k]
    pds=[]
    sts=[]
    for f in fls_pat:
        stdata = data.SpatialExpressionDataset.read_count_mat('/Users/joaquimteixeira/GSR_22_Tansey_Banerjee/teixeira_code/flipped_cohort_data/'+f)
        stddev_filtered = gene_filtering.select_top_genes_by_standard_deviation(
            stdata, n_gene=2000)

        spot_threshold_filtered = gene_filtering.filter_genes_by_spot_threshold(
            stddev_filtered, spot_threshold=0.95)

        stdata_filtered = gene_filtering.filter_ribosome_genes(spot_threshold_filtered)
        sts.append (stdata_filtered)
        df=pd.DataFrame(stdata_filtered.raw_counts)
        df.columns=stdata_filtered.gene_names
        pds.append (df)
    common_cols=pds[0].columns
    for df in pds:
        common_cols = list(set(common_cols).intersection(df.columns))
    pds[1][common_cols]
    pds2=[]
    best_n_components=n_comp[p]
    for df in pds:
        pds2.append(df[common_cols])
    lda=decomposition.LatentDirichletAllocation(n_components=best_n_components)
    pds=pd.concat(pds2)
    lda.fit (pds)
    trans=lda.transform (pds)
    n_genes=common_cols
    out=lda.__dict__
    with h5py.File('../teixeira_code/ldas_by_patient/'+pat+'.h5', 'w') as h5file:
        for key, item in out.items():
            if item is not None:
                if key is not 'random_state_':
                    h5file [key] = item
        h5file['transform']=np.array(trans)
        h5file['gene_names']=np.array(n_genes).astype(str)
    h5file.close()
    p=p+1
```
